```python
import pandas as pd
import numpy as np
import os

os.makedirs("../data/processed", exist_ok=True)
```

```python
prices = pd.read_csv("../data/raw/prices_raw.csv", index_col=0, parse_dates=True)
prices.head()
```

```python
returns = prices.pct_change().dropna()
returns.head()
```

```python
returns.shape
```

```python
factors = pd.read_csv("../data/raw/fama_french_factors.csv", parse_dates=["date"])
factors.set_index("date", inplace=True)
factors.head()
```

```python
factors_aligned = factors.reindex(returns.index).dropna()
factors_aligned.head()
```

```python
factors_aligned["Market"] = factors_aligned["Mkt-RF"] + factors_aligned["RF"]
```

```python
merged = returns.join(factors_aligned, how="inner")
merged.head()
```

```python
merged.shape
```

```python
returns.to_csv("../data/processed/returns.csv")
factors_aligned.to_csv("../data/processed/factors.csv")
merged.to_csv("../data/processed/merged_dataset.csv")

print("Saved:\n - returns.csv\n - factors.csv\n - merged_dataset.csv")
```
