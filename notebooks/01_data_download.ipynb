```python
import yfinance as yf
import pandas as pd
import numpy as np
import os

os.makedirs("../data/raw", exist_ok=True)
```

```python
tickers = [
    # Tech
    "AAPL","MSFT","AMZN","TSLA","GOOGL","META","NVDA","AVGO","ADBE","CRM",
    # Financials
    "JPM","BAC","WFC","GS","MS","C","V","MA","AXP","BLK",
    # Energy
    "XOM","CVX","COP","SLB","EOG",
    # Healthcare
    "JNJ","PFE","MRK","ABBV","LLY","UNH","TMO","ABT",
    # Consumer
    "PG","KO","PEP","MCD","HD","COST","DIS","NKE","SBUX",
    # Industrials
    "CAT","DE","GE","HON","BA"
]

tickers
```

```python
data = yf.download(
    tickers,
    start="2019-01-01",
    end="2025-01-01",
    auto_adjust=True
)["Close"]

data.head()
```

```python
data.to_csv("../data/raw/prices_raw.csv")
print("Saved to data/raw/prices_raw.csv")
```
