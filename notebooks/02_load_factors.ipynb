```python
import pandas as pd

file_path = "../data/F-F_Research_Data_5_Factors_2x3_daily.CSV"
df = pd.read_csv(file_path, skiprows=3)
df.head()
```

```python
df = df[df.iloc[:,0].str.isdigit()]
df.head()
df.tail()
```

```python
df = df.rename(columns={"Unnamed: 0": "date"})
```

```python
df["date"] = pd.to_datetime(df["date"], format="%Y%m%d")
```

```python
factor_cols = ["Mkt-RF", "SMB", "HML", "RMW", "CMA", "RF"]

df[factor_cols] = df[factor_cols].astype(float) / 100
```

```python
df.to_csv("../data/raw/fama_french_factors.csv", index=False)
print("Saved cleaned factors to data -> raw -> fama french factors")
```
