```python
import pandas as pd
import numpy as np

expected_returns = pd.read_csv("../data/results/expected_returns.csv", index_col=0).iloc[:, 0]
betas = pd.read_csv("../data/processed/factor_betas.csv", index_col=0)
merged = pd.read_csv("../data/processed/merged_dataset.csv", index_col=0, parse_dates=True)
```

```python
top_stocks = expected_returns.head(20)
top_stocks
```

```python
n = len(top_stocks)
weights_equal = np.ones(n) / n
weights_equal
```

```python
portfolio_return_equal = np.dot(weights_equal, top_stocks.values)
portfolio_return_equal
```

```python
factor_cols = ["Market", "SMB", "HML", "RMW", "CMA"]

beta_subset = betas.loc[top_stocks.index, factor_cols]

portfolio_exposure_equal = np.dot(weights_equal, beta_subset)
portfolio_exposure_equal
```

```python
bottom_stocks = expected_returns.tail(10)
bottom_stocks
```

```python
long_names = top_stocks.head(10).index
short_names = bottom_stocks.index

weights_longshort = pd.Series(0, index=expected_returns.index)

# long positions: +0.05 each
weights_longshort[long_names] = 0.1

# short positions: -0.05 each
weights_longshort[short_names] = -0.1

weights_longshort = weights_longshort[weights_longshort != 0]
weights_longshort
```

```python
ls_return = np.dot(weights_longshort, expected_returns.loc[weights_longshort.index])
ls_return
```

```python
beta_ls = betas.loc[weights_longshort.index, factor_cols]
portfolio_exposure_ls = np.dot(weights_longshort, beta_ls)
portfolio_exposure_ls
```
